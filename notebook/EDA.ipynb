## Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Reading data and visualising it

```python
train_df = pd.read_csv('../data/train.csv')
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=train_df, palette='magma')
plt.title('Survival Rate by Class and Sex')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.kdeplot(train_df[train_df['Survived'] == 1]['Age'], shade=True, color="green", label="Survived")
sns.kdeplot(train_df[train_df['Survived'] == 0]['Age'], shade=True, color="red", label="Not Survived")
plt.title('Age Distribution: Survived vs Not Survived')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(train_df['Fare'], bins=40, kde=True)
plt.title('Distribution of Fare')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
temp_df = train_df.copy()

temp_df['Sex'] = temp_df['Sex'].map({'male': 0, 'female': 1})

sns.heatmap(temp_df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation Matrix of Features')
plt.show()
```
